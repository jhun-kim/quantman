```python
import pandas as pd
```

```python
# data_load

df = pd.read_csv('./F-Score.csv', encoding='cp949')
df
```

```python
df.head()
```

```python
df.columns = ['name', 'code', "year", 'net_p_l', 'cash_flow', 'ROI', 'debt_r', 'PCR', 'stock', 'f_stock', 'sales_r', 'net_r', 'total_cap' ,'close']
```

```python
df.columns
```

```python
import numpy as np

df = df.fillna(0)
df['F_score'] = np.nan
df['F_score1'] = np.nan
df['F_score1'] = df['F_score1'].fillna(0)

company = list(df['name'].unique())
len(company)
df.columns
```

```python
#01 전년 당기순이익 흑자
df.loc[df['net_p_l'] > 0, 'F_score'] = 1
df.loc[df['net_p_l'] < 0, 'F_score'] = 0
```

```python
#02 전년 영업활동현금흐름 흑자
df.loc[df['cash_flow'] > 0, 'F_score'] += 1
df.loc[df['cash_flow'] < 0, 'F_score'] = 0
```

```python
#03 (현재 ROA > 전년 ROA)>>> 현재 ROI > 전년 ROI
for i in company :
    df_a = df[df['name'] == i]
    for j in range(len(df_a)-1):
        df.iloc[j+1,-1] = (df.iloc[j+1,5] - df.iloc[j,5])
df.loc[df['F_score1'] > 0, 'F_score'] += 1
```

```python
#04 현재 영업현금흐름 > 순이익 순이익 < 현금흐름
for i in company :
    df_a = df[df['name'] == i]
    for j in range(len(df_a)-1):
        df.iloc[j+1,-1] = (df.iloc[j,3] - df.iloc[j,4])
df.loc[df['F_score1'] > 0, 'F_score'] += 1
```

```python
#05 현재 부채비율 < 전년 부채비율
for i in company :
    df_a = df[df['name'] == i]
    for j in range(len(df_a)-1):
        df.iloc[j+1,-1] = (df.iloc[j+1,6] - df.iloc[j,6])
df.loc[df['F_score1'] < 0, 'F_score'] += 1
```

```python
#06 (현재 유동비율 > 전년 유동비율) >>> 금년 PCR < 전년 PCR
for i in company :
    df_a = df[df['name'] == i]
    for j in range(len(df_a)-1):
        df.iloc[j+1,-1] = (df.iloc[j+1,7] - df.iloc[j,7])
df.loc[df['F_score1'] < 0, 'F_score'] += 1
```

```python
#07 전년 신주발행이력 X
for i in company :
    df_a = df[df['name'] == i]
    for j in range(len(df_a)-1):
        df.iloc[j+1,-1] = (df.iloc[j+1,8] - df.iloc[j,8])
df.loc[df['F_score1'] <= 0, 'F_score'] += 1
```

```python
#08 (매출총이익률 전년대비증가) >>> 매출액증가율 < 순이익증가율
for i in company :
    df_a = df[df['name'] == i]
    for j in range(len(df_a)-1):
        df.iloc[j+1,-1] = (df.iloc[j,11] - df.iloc[j,10])
df.loc[df['F_score1'] > 0, 'F_score'] += 1
```

```python
#09 (자산회전율 전년대비증가) >>> 총자본회전율 전년대비증가
for i in company :
    df_a = df[df['name'] == i]
    for j in range(len(df_a)-1):
        df.iloc[j+1,-1] = (df.iloc[j+1,12] - df.iloc[j,12])
df.loc[df['F_score1'] > 0, 'F_score'] += 1
```
